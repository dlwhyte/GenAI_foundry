## Part 2: What Makes Something an Agent?

An agent is an LLM that can:

1. **Observe** — receive input from the world
2. **Think** — reason about what to do next (not just what to say)
3. **Act** — take an action that affects the world
4. **Loop** — repeat until the task is done

The key shift: **the LLM decides what happens next**, not just what to say.

```
Chatbot:  User → LLM → Response  (one shot)

Agent:    User → LLM → Action → Observation
                  ↑________________________|
                  (loop until task complete)
```

````python
# A chatbot does this (one step):
def chatbot(question):
    return f"LLM Response to: '{question}'"

print(chatbot('What is the capital of France?'))
print()

# An agent does this (multiple steps decided by the LLM):
def simulated_agent(goal):
    print(f'Goal received: {goal}')
    print('LLM decides: I need to search first...')
    print('  Action 1: web_search("top AI companies France 2025")')
    print('  Observation: Found: Mistral, Hugging Face, Snips...')
    print('LLM decides: I need more detail on each...')
    print('  Action 2: web_search("Mistral AI latest news")')
    print('  Observation: Mistral released new model in March 2025...')
    print('LLM decides: I have enough to write a summary.')
    print('  Action 3: generate_summary(...)')
    print('Task complete!')

simulated_agent('Research top AI companies in France and summarize their latest news')
````

## Part 3: The Four Building Blocks

### 🔧 1. Tools
Functions the agent can call — web search, calculators, databases, APIs.

### 🧠 2. Memory
- **Short-term**: the conversation so far (context window)
- **Long-term**: a vector store the agent can read/write (like your RAG pipeline!)
- **Episodic**: a log of past actions across sessions

### 📋 3. Planning (ReAct Pattern)
Reason + Act: the LLM reasons about what to do, acts, observes the result, then reasons again.

### ⚡ 4. Action
Execute a step, get back an observation, feed it into the next reasoning step.

````python
# A minimal agent loop — no API key needed!

def web_search(query):
    results = {
        'AI trends 2025': 'Top trends: multimodal models, agentic systems, on-device AI',
        'cost of GPT-4': 'GPT-4o costs $0.005 per 1K input tokens'
    }
    return results.get(query, f'3 relevant articles found for: {query}')

def calculator(expression):
    try:
        return str(eval(expression))
    except:
        return 'Error: invalid expression'

TOOLS = {'web_search': web_search, 'calculator': calculator}
agent_memory = []

def agent_step(thought, tool_name, tool_input):
    print(f'💭 Thought: {thought}')
    print(f'🔧 Action:  {tool_name}({repr(tool_input)})')
    result = TOOLS[tool_name](tool_input)
    agent_memory.append({'action': f'{tool_name}({tool_input})', 'observation': result})
    print(f'👁️  Observation: {result}')
    print()
    return result

print('=' * 60)
print('GOAL: How much would 500 API calls cost at GPT-4o pricing?')
print('=' * 60)
print()

pricing_info = agent_step(
    thought='I need to find current GPT-4o pricing first.',
    tool_name='web_search',
    tool_input='cost of GPT-4'
)

cost = agent_step(
    thought='Now calculate: 500 calls * avg 500 tokens = 250,000 tokens',
    tool_name='calculator',
    tool_input='500 * 500 * 0.005 / 1000'
)

print(f'✅ Final Answer: 500 API calls would cost approximately ${cost}')
print(f'📝 Agent used {len(agent_memory)} tools.')
````

## Part 4: Chatbot vs. Agent — Side by Side

| | Chatbot | Agent |
|---|---|---|
| **Input** | User message | User *goal* |
| **Output** | Text response | Completed task |
| **Tools** | None | Web, APIs, files, databases |
| **Memory** | This conversation only | Across sessions |
| **Who drives** | You drive every step | Agent drives, you supervise |
| **Cost** | Low (one LLM call) | Higher (multiple calls + tools) |
| **Failure modes** | Hallucination | Wrong tool, infinite loops, security risks |

````python
def should_use_agent(task):
    agent_signals = ['multiple steps','search','calculate','compare',
                     'monitor','schedule','send','fetch','look up','then']
    matches = [s for s in agent_signals if s in task.lower()]
    if matches:
        print(f'🤖 AGENT recommended — signals: {matches}')
    else:
        print(f'💬 CHATBOT is fine — single-step task')

tasks = [
    'What is the capital of France?',
    'Search for top AI papers this week, summarize them, then email me',
    'Explain what a transformer is',
    'Monitor my server logs and alert me if errors exceed 100 per minute',
    'Write a haiku about embeddings'
]

for task in tasks:
    print(f'Task: {task}')
    should_use_agent(task)
    print()
````
